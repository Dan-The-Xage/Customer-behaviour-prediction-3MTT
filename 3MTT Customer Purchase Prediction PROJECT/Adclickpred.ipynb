```python
# Selecting Features and Target
X = Adclick[['Age', 'EstimatedSalary', 'Gender']]
y = Adclick['Purchased']
```

```python
X
```

```python
y
```

```python
# Splitting into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Step 3: Model Training
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Calculate training score
y_train_pred = model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred)
print("Training Accuracy:", train_accuracy)

# Step 4: Model Evaluation
y_pred = model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", test_accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
# -------------------- Logistic Regression --------------------
print("\n--- Logistic Regression ---")
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)

# Training score
y_train_pred_lr = lr_model.predict(X_train)
train_accuracy_lr = accuracy_score(y_train, y_train_pred_lr)
print("Training Accuracy:", train_accuracy_lr)

# Evaluation
y_pred_lr = lr_model.predict(X_test)
test_accuracy_lr = accuracy_score(y_test, y_pred_lr)
print("Test Accuracy:", test_accuracy_lr)
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
```

```python
from sklearn.svm import SVC
# -------------------- Support Vector Machine (SVM) --------------------
print("\n--- Support Vector Machine (SVM) ---")
svm_model = SVC(random_state=42) # You might want to experiment with kernel='linear', 'rbf', 'poly', etc.
svm_model.fit(X_train, y_train)

# Training score
y_train_pred_svm = svm_model.predict(X_train)
train_accuracy_svm = accuracy_score(y_train, y_train_pred_svm)
print("Training Accuracy:", train_accuracy_svm)

# Evaluation
y_pred_svm = svm_model.predict(X_test)
test_accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Test Accuracy:", test_accuracy_svm)
print("Classification Report:\n", classification_report(y_test, y_pred_svm))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
```

```python
from sklearn.neighbors import KNeighborsClassifier
# -------------------- K-Nearest Neighbors (KNN) --------------------
print("\n--- K-Nearest Neighbors (KNN) ---")
knn_model = KNeighborsClassifier(n_neighbors=5) # You'll likely want to tune 'n_neighbors'
knn_model.fit(X_train, y_train)

# Training score
y_train_pred_knn = knn_model.predict(X_train)
train_accuracy_knn = accuracy_score(y_train, y_train_pred_knn)
print("Training Accuracy:", train_accuracy_knn)

# Evaluation
y_pred_knn = knn_model.predict(X_test)
test_accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("Test Accuracy:", test_accuracy_knn)
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
```

```python
from sklearn.tree import DecisionTreeClassifier
# -------------------- Decision Tree --------------------
print("\n--- Decision Tree ---")
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Training score
y_train_pred_dt = dt_model.predict(X_train)
train_accuracy_dt = accuracy_score(y_train, y_train_pred_dt)
print("Training Accuracy:", train_accuracy_dt)

# Evaluation
y_pred_dt = dt_model.predict(X_test)
test_accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Test Accuracy:", test_accuracy_dt)
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
# -------------------- Gradient Boosting --------------------
print("\n--- Gradient Boosting ---")
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_model.fit(X_train, y_train)

# Training score
y_train_pred_gb = gb_model.predict(X_train)
train_accuracy_gb = accuracy_score(y_train, y_train_pred_gb)
print("Training Accuracy:", train_accuracy_gb)

# Evaluation
y_pred_gb = gb_model.predict(X_test)
test_accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Test Accuracy:", test_accuracy_gb)
print("Classification Report:\n", classification_report(y_test, y_pred_gb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_gb))
```
